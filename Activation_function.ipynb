```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+(np.exp(-x)))
x=np.array([1,2,3])
result=sigmoid(x)
print(result)
plt.plot(x,result)
```

```python
def relu(x):
    return np.maximum(0, x)
x=np.array([1,2,3])
result=relu(x)
print(result)
plt.plot(x,result)
```

```python
def softmax(x):
    exps = np.exp(x - np.max(x))
    return exps / np.sum(exps)
x=np.array([1,2,3])
result= softmax(x)
print(result)
plt.plot(x,result)
```

```python
def tanh(x):
  return np.tanh(x)
x=np.array([1,2,3])
result=tanh(x)
print(result)
plt.plot(x,result)
```
